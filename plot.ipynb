```python
import matplotlib.pyplot as plt
import numpy as np

def plot_accuracy_vs_rounds_from_files(file_paths, title):
    plt.figure(figsize=(10, 6)) 

    colors = ['r', 'g', 'b', 'orange']

    for idx, file_path in enumerate(file_paths):
        with open(file_path, 'r') as file:
            data = [float(line.strip()) for line in file]

        rounds = list(range(1, len(data) + 1)) 

        plt.plot(rounds, data, linestyle='-', color=colors[idx % len(colors)], label=f'{file_paths[file_path]}')

    plt.title(title)
    plt.xlabel('Rounds')
    plt.ylabel('Accuracy')
    plt.yticks(np.arange(0, 1.0, 0.1))
    plt.legend(loc='best') 
    plt.grid(True)

    plt.show()
```

```python
file_paths = {
    'testes/0infected': 'Sem ataque',
    'testes/noProtectionCoordenate/10infected': "10% infectado",  
    'testes/noProtectionCoordenate/20infected': "20% infectado",
    'testes/noProtectionCoordenate/30infected': "30% infectado"  
}

plot_accuracy_vs_rounds_from_files(file_paths, 'Ataque coordenado em sistema desprotegido')
```
